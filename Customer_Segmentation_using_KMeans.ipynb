Assumption : We actively cannot guess the income of the customers so we will be removing those rows.

```python
df.dropna(inplace=True)  # removes the null rows
```

```python
# We need to change the Dt_Customer column since
sep_date = df['Dt_Customer'].str.split("-", n=3, expand=True)

df['day'] = sep_date[0].astype('int')
df['month'] = sep_date[1].astype('int')
df['year'] = sep_date[2].astype('int')
```

```python
df.nunique()
```

* Looking at this we can say that few of the columns like ['Z_CostContact','Z_Revenue'] are not useful because they give only contain 1 unique value.
* We also don't need ['Dt_Customer'] now since we have converted this column into 3 separate columns respectively.

```python
# Removing unnecessary columns
df.drop(columns=['Dt_Customer','Z_CostContact','Z_Revenue'], axis=1, inplace=True)
```

Now we are going to convert these object types into numerical columns so that we can feed it into our model.

```python
# converting categorical data into numerical
from sklearn.preprocessing import LabelEncoder

for col in df.columns:
    if df[col].dtype == 'object':
      le = LabelEncoder()
      df[col] = le.fit_transform(df[col])
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr()>0.8, annot=True, cbar=False)
plt.show()
```

```python
#standardizing the values of each column using Standard Scaler

scaler = StandardScaler()
data = scaler.fit_transform(df)
```

```python
from sklearn.manifold import TSNE

model = TSNE(n_components = 2, random_state = 0)
tsne_data = model.fit_transform(df)

plt.figure(figsize=(7,7))
plt.scatter(tsne_data[:, 0], tsne_data[:, 1])
plt.show()
```

From this graph we can interpret there are clearly some kind of cluster one way or the other so let's move futher with the KMeans.

```python
error = []
for n_cluster in range(1, 21):
  model = KMeans(init = 'k-means++', n_clusters = n_cluster,
                 max_iter = 500, random_state=42)
  model.fit(df)
  error.append(model.inertia_)
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x= range(1,21), y=error)
sns.scatterplot(x=range(1,21), y=error)
plt.show()
```

With this we can say that the k=6 is the most optimal number of clusters that should be made as k=6 the value of the inertia is not decreasing drastically.

```python
model = KMeans(init = 'k-means++', n_clusters=5,
               max_iter = 500, random_state=22)
segments = model.fit_predict(df)
```

```python
plt.figure(figsize=(7,7))
df_tsne = pd.DataFrame({"x": tsne_data[:,0],'y': tsne_data[:,1], 'segment' : segments})

sns.scatterplot(x='x', y='y', hue='segment', data = df_tsne)
plt.show()
```

From this we can see that there is a clear difference between the types of cutomers.
